### $\color{darkblue}{\text{Hypothesis testing}}$ 

Let's answer the very first question we asked ourselves: how can we quantify our level of confidence in stating that this difference is $\textbf{NOT}$ given by low sample-size effect? 

The answer is incredibly fascinating in its simplicity and effectiveness. The idea is the following: 
* *Create new datasets by permuting randomly the labels* 
* *Compute the differences in this new randomly drawn datasets*
* *See how many times the differences are higher than the experimental one*

We must create equivalent datasets made of a randomized group of 541 people (Group A) and another made of 774 (Group B) corresponding to what previously were respectively brexiteers and remainers.

````python
# Create a list with all ages

global_stat = list(brexiteers['age']) +  list(remainers['age'])

print(len(global_stat))
````

Import random package for computing random permutations

````python
import random 
````

Let's create 10000 new shuffled datasets and let's keep track when one difference in average ages is greater than the observed experimental one

Note: we create a fake scenario to verify our theory.
We suppose the theory is wrong, thus we create random fake datsets and see if the result is similar to the real data.

````python
trials = 10000  ; count = 0

for dumb in range(trials):
    # Compute shuffled ages
    random.shuffle(global_stat)
    # Compute average of Group A
    avg_A = sum(global_stat[:541])/541 
    # Compute average of Group B
    avg_B = sum(global_stat[541:])/774 
    # Compute new shuffled difference 
    diff = avg_A - avg_B
    if diff > dataset_property:
        print('ciao')
        count +=1
````

...and we finally arrive to the result

````python
print(f"The average age of Group A was less than the one of Group B {count} times") 
````

We could safely $\textbf{reject}$ the null hypothesis 

Note: we reject the null hypothesis because, in creating a scenario where the null hypothesis is correct, only 2 times out 10000 the fake average age difference is greater than the measured one.
Essentialy, we are veryfying that we do not get this result "by chance". As the result similar to the one of the real world happens only two times we can guarantee that this does not happen by chance.

## $\color{yellow}{\text{Bootstrap}}$ 

**Bootstrap** is a resampling strategy with replacement that requires no assumptions about the data distribution. 

It allows us to make inferences about the population statistics when we only have a finite number of samples. In the following we will see how to use it to evaluate the **confidence interval** and the **variance** of an estimator, but it has many other applications.

Even when we only have one sample, the bootstrap method provides a good enough approximation of the true population statistics.

##### Compute the following conditional probabilities:

* $P(Survived=1\;|\; Sex = male\;,\; Pclass = 1)$
* $P(Survived=1\;|\; Sex = male\;,\; Pclass = 2)$
* $P(Survived=1\;|\; Sex = male\;,\; Pclass = 3)$

After finding the values of $Pclass$ corresponding to the highest and the lowest value of survival rate, test the hypothesis that the difference is due to the limited number of data.

````python
Pclass = [1, 2, 3]
probs_list = list()
for c in Pclass:
    males_inClass = males[males['Pclass'] == c]
    survived_inClass = males_inClass[males_inClass['Survived'] == 1]
    #print(f"tot: {len(males_inClass)}, survived: {len(survived_inClass)}, class: {c}")
    probability = len(survived_inClass) / len(males_inClass)
    probs_list.append(probability)
    print(f"P(survived | male, {c} class) = {probability}")
````

1st class has the highest survival rate. 3rd class the lowest

To see if the difference is due to the limited amount of data I use the null hypothesis.

````python
def compute_survivalRate(survived_inClass, males_inClass):
    probability = len(survived_inClass) / len(males_inClass)
    return probability
````

````python
import random

males_1class =  males[males['Pclass'] == 1]
survival_1class = males_1class['Survived']

males_3class = males[males['Pclass'] == 3]
survival_3class = males_3class['Survived']

global_stat = list(survival_1class) + list(survival_3class)

dataset_property = probs_list[0] - probs_list[2] # difference between survival rate of 1 class and 3 class
trials = 10000  ; count = 0
diff_list = list()
for dumb in range(trials):
    # Compute shuffled ages
    random.shuffle(global_stat)
    # Compute average of Group A
    survivalRate_A = sum(global_stat[:len(survival_1class)])/ len(survival_1class)
    # Compute average of Group B
    survivalRate_B = sum(global_stat[len(survival_1class):])/ len(survival_3class)
    # Compute new shuffled difference 
    diff = survivalRate_A - survivalRate_B
    diff_list.append(diff)
    if diff > dataset_property:
        count +=1
        
print(f"There are {count} out of 10000 differences from the null hypothesis that are greater than the real difference")
````

There are zero cases out of 10000 random datasets where the same relation occurs, thus we can safely assume it is not due to the limited dataset.

I also want to see if the statistical inference may hold also with a larger population, thus I apply bootstrapping.
I compute different values of survival rate for different resampled datasets of the original one.
Then I compute the mean and variance of the survival rate distribution and the confidence interval with a 95% percentile.

````python
reps = 10000

# first class

males_1class =  males[males['Pclass'] == 1]
n = len(males_1class)
survival_1class =  pd.Series.to_numpy(males_1class['Survived'])
boot_survived_1class = np.random.choice(survival_1class, (reps, n))

males_3class = males[males['Pclass'] == 3]
n = len(males_3class)
survival_3class = pd.Series.to_numpy(males_3class['Survived'])
boot_survived_3class = np.random.choice(survival_3class, (reps, n))
````

````python

def compute_boot_survivalRate(boot_survived):
    survival_rates = list()
    for set in boot_survived:
        survived = np.sum(set)
        probability = survived / len(set)
        survival_rates.append(probability)
    
    return np.array(survival_rates)

boot_survivalRate_1class = compute_boot_survivalRate(boot_survived_1class)
boot_survivalRate_3class = compute_boot_survivalRate(boot_survived_3class)
````

````python
mean_1class = boot_survivalRate_1class.mean()
mean_3class = boot_survivalRate_3class.mean()

print(f"Mean survival rate 1st class: {mean_1class}\nMean survival rate 3rd class: {mean_3class}")
````

````python
std_1class = boot_survivalRate_1class.std()
std_3class = boot_survivalRate_3class.std()

print(f"Std deviation survival rate 1st class: {std_1class}\nStd deviation survival rate 3rd class: {std_3class}")
````

The two means are reasonable as they are very close to the survival rate of the original set.
Also values of std deviation seem coherent, as they are small, indicating that the obtained distributions is concentrated close to the mean and they do not overlap.
Let's verify that also with confidence intervals.

````python
boot_survivalRate_1class_CI = np.percentile(boot_survivalRate_1class, [2.5, 97.5])
boot_survivalRate_3class_CI = np.percentile(boot_survivalRate_3class, [2.5, 97.5])
````

````python
fig = plt.figure(figsize=(9,5))
ax = fig.add_subplot(111)
plt.title("Survival rate distribution", size = 16)
ax.hist(boot_survivalRate_1class, color ='darkred', label = f"1st class")
plt.axvline(boot_survivalRate_1class_CI[0],color = 'red',linewidth=2)
plt.axvline(boot_survivalRate_1class_CI[1],color = 'red',linewidth=2)
ax.hist(boot_survivalRate_3class, color ='darkblue', label = f"3rd class")
plt.axvline(boot_survivalRate_3class_CI[0],color = 'blue',linewidth=2)
plt.axvline(boot_survivalRate_3class_CI[1],color = 'blue',linewidth=2)
ax.set(xlabel ="Survival rate" , ylabel = "#(occurrences)")
ax.legend();
````
